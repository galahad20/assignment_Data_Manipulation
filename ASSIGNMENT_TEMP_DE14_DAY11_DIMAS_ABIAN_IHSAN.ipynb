## Group By dan Agregasi


### Hitung total penjualan (total_amount) per category

```python
order.groupby('category')['total_amount'].sum()
```

### Hitung jumlah order dan rata-rata total_amount per city

```python
order.groupby('city').agg({
    'order_id' : 'count',
    'total_amount' : 'mean'
}).rename(columns={
    'order_id' : 'jumlah_order',
    'total_amount' : 'rata_rata_total_amount'
})
```

### Tentukan top 5 customer dengan total pembelian tertinggi

```python
order.groupby('customer_name').agg({
    'total_amount' : 'sum'
}).sort_values('total_amount', ascending=False).head()
```

## Join/Merge Data

### Gabungkan tabel orders dengan products berdasarkan product_id

```python
order_with_prod = order.merge(prod, on='product_id', how='inner')
order_with_prod.head()
```

### Gabungkan tabel orders dengan customers berdasarkan customer_id

```python
order_with_cust = order.merge(cust, on='customer_id', how='inner')
order_with_cust.head()
```

### Hitung profit per order (gunakan unit_price dan unit_cost dari products)

Disini saya simpan nilai profit ke dalam kolom baru agar bisa digunakan lagi nantinya

```python
#perhitungan profit
order_with_prod['profit'] = ((order_with_prod['unit_price_y'] * order_with_prod['quantity']) * (1 - order_with_prod['discount_pct']/100))  - (order_with_prod['unit_cost'] * order_with_prod['quantity'])

#filtering with group by
order_with_prod.groupby('order_id')['profit'].sum().reset_index()
```

```python
#detail order_per_profit
order_with_prod.groupby('order_id').agg({
    'product_name_y' : 'first',
    'total_amount' : 'first',
    'discount_pct' : 'first',
    'quantity' : 'sum',
    'unit_price_y' : 'first',
    'unit_cost' : 'first',
    'profit' : 'sum'
}).reset_index().head()
```

### Identifikasi hari dalam seminggu dengan penjualan tertinggi

Jujur soalnya agak ambigu banyak penafsiran berbeda seperti

- apakah identifikasi hari dalam seminggu dengan penjualan tertinggi dan itu diulang lagi untuk minggu lainnya dan ditampilkan per minggu

- Atau identifikasi hari dalam seminggu dengan penjualan tertinggi. Kemudian jumlahkan dengan hari yang sama di minggu lain sehingga dapat hasilnya dan kemudian bandingkan dengan hari lain dalam seminggu. Sehingg didapatkan total penjualan tertinggi itu pada hari A misalnya jika dibandingkan dengan total hari lainnya.

- Atau bahkan identifikasi seluruh hari dalam seminggu di dataset itu dan hanya tampilkan hari dengan penjualan tertingginya saja, meskipun hari itu biasanya tidak tinggi penjualannya di minggu lainnya namun selama masih penjualan tertinggi (misal tanggal kembar) itu yang ditampilkan.

Namun saya sampati dengan kesimpulan ke-2 bahwa identifikasi hari yang akumulasi atau tren keseluruhan dalam seminggu di tiap minggunya memilki total penjualan tertinggi.

```python
#membuat kolom day dari order date
order['day'] = order['order_date'].dt.day_name()
```

```python
order.head()
```

```python
#group by day dan sort by day dengan penjualan tertinggi
order.groupby('day')['total_amount'].sum().reset_index().sort_values(by='total_amount', ascending=False)
```

Dari hasil di atas akan menampilkan tren yang mana dari data yang diberikan, monday menjadi hari dengan total penjualan tertinggi.
